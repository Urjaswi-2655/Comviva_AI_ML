```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans

# Assuming 'data' is your DataFrame containing the relevant columns
selected_columns = ['user_segment', 'revenue_range']

# Step 1: Use column names to select specific columns
X = data[selected_columns].values

# Step 2: Impute NaN values using SimpleImputer with mean strategy
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Step 3: Now, use the filled data with KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_imputed)

    wcss.append(kmeans.inertia_)
```

### Training the model

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans

# Assuming 'data' is your DataFrame containing the relevant columns
selected_columns = ['user_segment', 'revenue_range']

# Step 1: Use column names to select specific columns
X = data[selected_columns].values

# Step 2: Impute NaN values using SimpleImputer with mean strategy
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Step 3: Filter out rows containing NaN values (if any)
X_cleaned = X_imputed[~pd.isna(X_imputed).any(axis=1)]

# Step 4: Apply KMeans to the cleaned data
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0)
Y = kmeans.fit_predict(X_cleaned)

print(Y)
```
